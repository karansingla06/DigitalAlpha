```python
# Required Python Machine learning Packages
import pandas as pd
import numpy as np
# For preprocessing the data
from sklearn.preprocessing import Imputer
from sklearn import preprocessing
# To split the dataset into train and test datasets
from sklearn.model_selection import train_test_split
# To model the Gaussian Navie Bayes classifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB
```

```python
df=pd.read_csv("C:/Users/user/Desktop/adult.csv")
```

```python
df.isnull().sum()
```

```python
df=df.replace(" ?",np.NaN)
df
```

```python
df = df.apply(lambda x:x.fillna(x.value_counts().index[0]))
df
```

```python
le = preprocessing.LabelEncoder()
workclass_cat = le.fit_transform(df.workclass)
education_cat = le.fit_transform(df.education)
marital_cat   = le.fit_transform(df['marital-status'])
occupation_cat = le.fit_transform(df.occupation)
relationship_cat = le.fit_transform(df.relationship)
race_cat = le.fit_transform(df.race)
sex_cat = le.fit_transform(df.sex)
native_country_cat = le.fit_transform(df['native-country'])
```

```python
df2=df
df2
```

```python
df2.describe()
```

```python
df2['workclass_cat'] = workclass_cat
df2['education_cat'] = education_cat
df2['marital_cat'] = marital_cat
df2['occupation_cat'] = occupation_cat
df2['relationship_cat'] = relationship_cat
df2['race_cat'] = race_cat
df2['sex_cat'] = sex_cat
df2['native_country_cat'] = native_country_cat
```

df2.head()

```python
df2.head()
```

```python
dummy_fields = ['workclass', 'education', 'marital-status', 
                  'occupation', 'relationship', 'race',
                  'sex', 'native-country']
df2 = df2.drop(dummy_fields, axis = 1)
```

```python
df2.head()
```

```python
#Standardization of data

num_features = ['age', 'workclass_cat', 'fnlwgt', 'education_cat', 'education-num',
                'marital_cat', 'occupation_cat', 'relationship_cat', 'race_cat',
                'sex_cat', 'capital-gain', 'capital-loss', 'hours-per-week',
                'native_country_cat']
 
scaled_features = {}
for each in num_features:
    mean, std = df2[each].mean(), df2[each].std()
    scaled_features[each] = [mean, std]
    df2.loc[:, each] = (df2.loc[:,each] - mean)/std
df2.head()
```

df2.head()

```python
df2 = df2.reindex(['age', 'workclass_cat', 'fnlwgt', 'education_cat',
                                    'education-num', 'marital_cat', 'occupation_cat',
                                    'relationship_cat', 'race_cat', 'sex_cat', 'capital-gain',
                                    'capital-loss', 'hours-per-week', 'native_country_cat', 
                                    'income'], axis= 1)
```

```python
X= df2.values[:,:14]
y = df2.values[:,14]
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 10)
```

```python
clf1 = GaussianNB()
clf1.fit(X_train, y_train)
```

```python
print("Accuracy is : {:0.4f}".format(clf1.score(X_test,y_test)*100))
```
